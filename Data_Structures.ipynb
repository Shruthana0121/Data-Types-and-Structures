8. What is a hash table, and how does it relate to dictionaries in Python?

--> A hash table is a data structure that stores data in key-value pairs and uses a hash function to quickly find values based on their keys.

Python dictionaries are implemented using hash tables.

10. Explain why strings are immutable in Python?

--> Strings are used everywhere in a program — variable names, file paths, function names, messages, etc.

If strings were mutable:

Python would need to re-calculate hash values each time a string changes.This would slow down dictionary operations

By making strings immutable:

Python can cache and reuse strings safely. String operations become faster.

11. What advantages do dictionaries offer over lists for certain tasks?

--> Faster lookups: Dictionaries use hash tables, so finding a value by its key is instant.

Key-Value Pair Storage: Dictionaries let you store data in a name → value format.

No Need to Remember Index Positions: With lists, you must remember where each item is stored.

Better for Structured Data: If you want to group attributes of an item, dictionaries are perfect.

12. Describe a scenario where using a tuple would be preferable over a list?

--> Use a tuple when you need data that should remain constant, faster performance, or a hashable structure (e.g., dictionary keys).

13. How do sets handle duplicate values in Python?

--> Sets don't allow duplicates.

14. How does the “in” keyword work differently for lists and dictionaries?

-->
“in” with Lists → Checks for VALUES: When used with a list, in checks whether an element/value is present anywhere in the list.

“in” with Dictionaries → Checks for KEYS: When used with a dictionary, in checks only the keys, not the values.

15. Can you modify the elements of a tuple? Explain why or why not.

--> No, you cannot modify the elements of a tuple in Python. Tuples are immutable, meaning once they are created, their elements cannot be changed, added, or removed.

16. What is a nested dictionary, and give an example of its use case?

-->A nested dictionary is a dictionary inside another dictionary.
It allows you to store structured, hierarchical data — like a dictionary with multiple levels.


17. Describe the time complexity of accessing elements in a dictionary.

--> Accessing elements in a Python dictionary is very fast because dictionaries use hash tables internally. Average time complexity: O(1)


18. In what situations are lists preferred over dictionaries?

-->
Use a list when:
Order matters,
duplicates are allowed,
you access items by position,
you just need a simple ordered collection.


Use a dictionary when:
you need key-value, relationships, fast lookups by key, structured data.


19. Why are dictionaries considered unordered, and how does that affect data retrieval?

-> Dictionaries in Python are considered unordered (conceptually) because they do not store elements based on a fixed positional index like lists do. Instead, they store data based on key-value pairs, and the internal storage is optimized for fast lookup using hashing—not for maintaining order.
